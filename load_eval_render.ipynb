```python
# Rendering only directly supported in jupyter notebooks.
from IPython.display import HTML
from brax.io import html
from brax import envs
from brax import jumpy as jp
import jax
import jax.numpy as jnp
from turbo import Turbo1
from brax_caller import BraxCaller
import numpy as np
import time
```

```python
# Only need to change this cell contents to load and visualize some policy
fn = "best_2_6187.1162109375_halfcheetah_10_10.npy"
env_name = "halfcheetah"
arch1 = 10
arch2 = 10
```

```python
brx_env = BraxCaller(env_name, arch1, arch2, 512, 8, "min", np.random.randint(1e6))
x = np.load(fn)
x_fdict = brx_env.dict_builder(x)
env = envs.create(env_name=env_name)
state = env.reset(rng=jp.random_prngkey(seed=np.random.randint(1e6)))
jit_env_step = jax.jit(env.step)
```

```python
rollout = []
state = env.reset(rng=jp.random_prngkey(seed=np.random.randint(1e6)))
rollout.append(state)
for i in range(1000):
    actions = brx_env.actor.apply(x_fdict, state.obs)
    state = jit_env_step(state, actions)
    print(state.reward, state.done)
    rollout.append(state)    
```

```python
HTML(html.render(env.sys, [s.qp for s in rollout]))
```
